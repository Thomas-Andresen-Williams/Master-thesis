```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import numpy as np
import matplotlib.pyplot as plt

# Define model complexity
model_complexity = np.linspace(0, 10, 100)

# Define bias and variance such that they cross at a specific point
bias = (10 - model_complexity) ** 2 / 100  # Bias decreases with complexity
variance = model_complexity ** 2 / 100     # Variance increases with complexity
noise = 0.02  # Fixed noise in the data

# Total error: Sum of bias^2, variance, and noise
total_error = bias + variance + noise

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(model_complexity, bias, label="Squared Bias", color="blue", linewidth=2)
plt.plot(model_complexity, variance, label="Variance", color="red", linewidth=2)
plt.plot(model_complexity, total_error, label="Total Error", color="green", linestyle="--", linewidth=2)

# Add labels and legend
plt.title("Bias-Variance Trade-off", fontsize=16)
plt.xlabel("Model Complexity", fontsize=14)
plt.ylabel("Test Error", fontsize=14)
plt.axvline(model_complexity[np.argmin(total_error)], color="purple", linestyle=":", label="Optimal Complexity")
plt.legend(fontsize=8)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Simulated feature names and importance scores
features = ['Feature A', 'Feature B', 'Feature C', 'Feature D', 'Feature E']
importance_scores = [0.35, 0.25, 0.15, 0.13, 0.12]

# Sort features by importance
sorted_idx = np.argsort(importance_scores)[::-1]
sorted_features = [features[i] for i in sorted_idx]
sorted_importance_scores = [importance_scores[i] for i in sorted_idx]

# Plot
plt.figure(figsize=(8, 6))
bars = plt.barh(sorted_features, sorted_importance_scores, color='skyblue', edgecolor='black')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest importance at the top
plt.xlabel("Importance Score", fontsize=12)
plt.title("Variable Importance Plot", fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Annotate bar values
for bar in bars:
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height() / 2, 
             f'{bar.get_width():.2f}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
